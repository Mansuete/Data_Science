```python
bottomLeft = (49.78, 23.9)
bottomRight = (49.78, 24.1483)
topLeft = (49.89983154, 23.9)
topRight = (49.89983154, 24.1483)

cols = np.linspace(bottomLeft[1], bottomRight[1], num=10)
rows = np.linspace(bottomLeft[0], topLeft[0], num=9)
new_data['col'] = np.searchsorted(cols, new_data.pickup_lng)
new_data['row'] = np.searchsorted(rows, new_data.pickup_lat)
```

```python
lat, lng = 49.839570, 23.994951

tmp_col = np.searchsorted(cols, lng)
tmp_row = np.searchsorted(rows, lat)

cell = str(tmp_col) + '_' + str(tmp_row)
le.transform(np.array([cell]).reshape((1, 1)))
```

```python
new_data['index_cell'] = new_data.col.map(str) + '_'+ new_data.row.map(str)
```

```python
new_data.drop(['col','row'], axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(new_data.index_cell)
```

```python
new_data.index_cell = le.transform(new_data.index_cell)
```

```python
data = new_data
```

```python
data['created_at'] = pd.to_datetime(data['created_at'])
data['hour'] = data['created_at'].dt.hour
data['day_of_month'] = data['created_at'].dt.day
data['week_day'] = data['created_at'].dt.weekday
data['month'] = data['created_at'].dt.month
```

```python
data
```

```python
tmp = data[['index_cell', 'user_id']].groupby(['index_cell']).apply(lambda x: x.user_id.nunique())
```

```python
tmp.to_csv('unique_users_per_area.csv')
```

```python
tmp = data[['index_cell', 'driver_id']].groupby(['index_cell']).apply(lambda x: x.driver_id.nunique())
```

```python
tmp.to_csv('unique_drivers_per_area.csv')
```

```python
tmp = data[['index_cell', 'ride_distance']].groupby(['index_cell']).mean()
```

```python
tmp.to_csv('mean_distance_per_area.csv')
```

```python
tmp = data[['index_cell', 'canceled_by_client', 'canceled_by_driver']].groupby(['index_cell']).mean()

tmp.to_csv('canceled_ratio.csv')
```

```python
tmp = data[['index_cell', 'ride_to_suburb']].groupby(['index_cell']).sum()

tmp.to_csv('suburb.csv')
```
