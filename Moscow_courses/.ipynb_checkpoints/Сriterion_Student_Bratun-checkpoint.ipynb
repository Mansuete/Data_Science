```python
import numpy as np
import pandas as pd
import scipy
from statsmodels.stats.weightstats import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r'd:\DownLoads\ADHD.txt', sep=' ', header=0)
data.columns = ['Placebo','Methyl']
```

```python
data.plot.scatter('Placebo','Methyl',c='r',s=30)
plt.grid()
plt.xlim(20,80)
plt.ylim(20,80)
plt.plot(range(100))
```

```python
scipy.stats.ttest_1samp(data.Placebo, 50.0)
```

```python
zconfint(data.Placebo)
```

```python
stats.probplot(data.Placebo,dist='norm',plot=plt)
```

```python
stats.probplot(data.Methyl,dist='norm',plot=plt)
```

```python
stats.shapiro(data.Placebo)
```

```python
stats.shapiro(data.Methyl)
```

```python
scipy.stats.ttest_ind(data.Placebo,data.Methyl,equal_var=False)
```

```python
cm = CompareMeans(DescrStatsW(data.Methyl),DescrStatsW(data.Placebo))
```

```python
cm.tconfint_diff(alpha=0.05)
```

```python
stats.probplot(data.Methyl - data.Placebo,dist='norm',plot=plt)
```

```python
stats.shapiro(data.Methyl - data.Placebo)
```

```python
stats.ttest_rel(data.Methyl,data.Placebo)
```

```python
ss = DescrStatsW(data.Methyl - data.Placebo)
```

```python
ss.tconfint_mean()
```
