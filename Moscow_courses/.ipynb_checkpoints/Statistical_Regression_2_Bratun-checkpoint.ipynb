```python
import statsmodels
import scipy as sc
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
from statsmodels.graphics.regressionplots import plot_leverage_resid2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
raw = pd.read_csv(r'd:\DownLoads\beauty.csv', sep=';', index_col=False)
raw.head()
```

```python
pd.tools.plotting.scatter_matrix(raw[['wage','exper','educ','looks']],alpha=0.2,figsize=(15,15),diagonal='hist')
```

```python
print(raw.union.value_counts())
print(raw.goodhlth.value_counts())
print(raw.black.value_counts())
print(raw.female.value_counts())
print(raw.married.value_counts())
print(raw.service.value_counts())
```

```python
data = raw
```

```python
data['wage'].hist()
```

```python
np.log(data['wage']).hist()
```

```python
data = data[data['wage'] < 77]
```

```python
data.shape
```

```python
data['looks'].hist()
```

```python
data['belongawg'] = data['looks'].apply(lambda x : 1 if x < 3 else 0)
data['aboveawg'] = data['looks'].apply(lambda x : 1 if x > 3 else 0)
```

```python
data.drop('looks', axis=1, inplace=True)
```

```python
data
```

```python
m1 = smf.ols('wage ~ exper + union + goodhlth + black + female + married + service + educ + belongawg + aboveawg', data=data)
```

```python
fitted = m1.fit()
```

```python
print(fitted.summary())
```

```python
sc.stats.probplot(fitted.resid,dist='norm',plot=plt)
# np.log(fitted.resid).hist()
```

```python
m2 = smf.ols('np.log(wage) ~ exper + union + goodhlth + black + female + married + service + educ + belongawg + aboveawg', data=data)
fitted = m2.fit()
print(fitted.summary())
sc.stats.probplot(fitted.resid, dist='norm', plot=plt)
```

```python
fitted.resid
```

```python
plt.scatter(data['exper'], fitted.resid)
```

```python
m3 = smf.ols('np.log(wage) ~ exper + np.power(exper,2) + union + goodhlth + black + female + married + service + educ + belongawg + aboveawg', data=data)
fitted = m3.fit()
print(fitted.summary())
sc.stats.probplot(fitted.resid, dist='norm', plot=plt)
```

```python
sms.het_breushpagan(fitted.resid,fitted.model.exog)[1]
```

```python
m4 = smf.ols('np.log(wage) ~ exper + np.power(exper,2) + union + goodhlth + black + female + married + service + educ + belongawg + aboveawg', data=data)
fitted = m4.fit(cov_type='HC1')
print(fitted.summary())
sc.stats.probplot(fitted.resid, dist='norm', plot=plt)
```

```python
m5 = smf.ols('np.log(wage) ~ exper + np.power(exper,2) + union + female + service + educ + belongawg + aboveawg', data=data)
fitted = m5.fit(cov_type='HC1')
print(fitted.summary())
sc.stats.probplot(fitted.resid, dist='norm', plot=plt)
```

```python
m4.fit().compare_f_test(m5.fit())
```

```python
plot_leverage_resid2(fitted)
```
