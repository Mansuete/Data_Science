```python
from sklearn.pipeline import Pipeline
import sklearn.pipeline
```

```python
pipeline = Pipeline(steps=[('scaling',scaler),('regression',regressor)])
```

```python
pipeline.fit(train_data,train_labels)
```

```python
metrics.mean_absolute_error(test_labels,pipeline.predict(test_data))
```

```python
parametrs = {
    'regression__loss':['huber','epsilon_insensitive','squared_loss'],
    'regression__n_iter':[3,5,10,50],
    'regression__penalty':['l1','l2','none'],
    'regression__alpha':[0.0001,0.01],
    'scaling__with_mean':[0,1]
}
```

```python
grid_cv = grid_search.GridSearchCV(pipeline,parametrs,scoring='mean_absolute_error',cv=4)
```

```python
%%time
grid_cv.fit(train_data,train_labels)
```

```python
grid_cv.best_score_
```

```python
grid_cv.best_params_
```

```python
metrics.mean_absolute_error(test_labels,grid_cv.best_estimator_.predict(test_data))
```

```python
numeric_data_columns = ['temp','atemp','humidify','windspeed','hour']
```

```python
numeric_data_indicies = np.array([(column in numeric_data_columns) for column in train_data.columns], dtype=bool)
```

```python
from sklearn.preprocessing import FunctionTransformer,OneHotEncoder
from sklearn.pipeline import FeatureUnion
```

```python
estimator = Pipeline(steps = [       
    ('feature_processing', FeatureUnion(transformer_list = [        
            #binary
            ('binary_variables_processing', FunctionTransformer(lambda data: data[:, binary_data_indices])), 
                    
            #numeric
            ('numeric_variables_processing', Pipeline(steps = [
                ('selecting', FunctionTransformer(lambda data: data[:, numeric_data_indicies])),
                ('scaling', StandardScaler(with_mean = 0))            
                        ])),
        
            #categorical
            ('categorical_variables_processing', Pipeline(steps = [
                ('selecting', FunctionTransformer(lambda data: data[:, catrgorical_data_indies])),
                ('hot_encoding', OneHotEncoder(handle_unknown = 'ignore'))            
                        ])),
        ])),
    ('model_fitting', regressor)
    ]
)
```

```python
estimator.fit(train_data,train_labels)
```

```python
metrics.mean_absolute_error(test_labels,estimator.predict(test_data))
```
