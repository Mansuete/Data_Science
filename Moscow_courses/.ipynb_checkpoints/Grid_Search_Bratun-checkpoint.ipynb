```python
from sklearn import cross_validation,datasets,grid_search,linear_model,metrics
import numpy as np
import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
train_data,test_data,train_labels,test_labels = cross_validation.train_test_split(iris.data,iris.target,test_size=0.3,random_state=0)
```

```python
classifier = linear_model.SGDClassifier(random_state=0)
```

```python
classifier.get_params()
```

```python
parametr_grid = {
    'loss':['hinge','log','squared_hinge','squared_loss'],
    'penalty':['l1','l2'],
    'n_iter':[5,6,7,8,9],
    'alpha':np.linspace(0.0001,0.001,num=5),
}
```

```python
cv = cross_validation.StratifiedShuffleSplit(train_labels,n_iter=10,test_size=0.2,random_state=0)
```

```python
grid_cv = grid_search.GridSearchCV(classifier,param_grid=parametr_grid,scoring='accuracy',cv=cv)
```

```python
%%time
grid_cv.fit(train_data,train_labels)
```

```python
grid_cv.best_estimator_
```

```python
grid_cv.best_score_
```

```python
grid_cv.best_params_
```

```python
grid_cv.grid_scores_
```

```python
random_grid_cv = grid_search.RandomizedSearchCV(classifier,parametr_grid,scoring='accuracy',cv=cv,n_iter=20,random_state=0)
```

```python
%%time
random_grid_cv.fit(train_data,train_labels)
```

```python
random_grid_cv.best_score_
```

```python
random_grid_cv.best_params_
```
